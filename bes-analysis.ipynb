```python
import beams
```

# Load pinis 1-4

```python
pinis = [beams.HeatingBeam(pini=pini, bvoltage=60e3, species='protium') for pini in [1,2,3,4]]
```

# Field alignment and Doppler shift along beam axis

```python
for ipini in [0,1]:
    pinis[ipini].plot_onaxis()
```

# Field alignment and Doppler shift in beam vertical plane

```python
for ipini in [0,1]:
    pinis[ipini].plot_vertical_plane('A21-lolo')
```

# Beam-weighted sightline calculations

```python
r_obs, z_obs = 5.8, -0.4
for ipini in [1]:
    pinis[ipini].plot_sightline(port='A21-lolo', r_obs=r_obs, z_obs=z_obs)
```
